```python
import keras
import tensorflow as tf
```

```python
from keras.layers import Dense, BatchNormalization, Conv2D, Flatten, Dropout
from keras.applications import InceptionV3
from keras.applications.resnet50 import ResNet50
from keras.utils import to_categorical
from keras.preprocessing.image import ImageDataGenerator,img_to_array
from keras.models import Model, load_model, save_model, Sequential
```

```python
import os
from glob import glob
dataset = os.path.join("Dataset")
train_dataset = os.path.join(dataset, "train")
test_dataset = os.path.join(dataset, "test")
```

```python
train_folders = glob('dataset/train/*')
len(train_folders)
```

```python
test_folders = glob('dataset/test/*')
len(test_folders)
```

```python
#Using Transfer learning
img_input = keras.layers.Input(shape=(150, 150,1))
img_conc = keras.layers.Concatenate()([img_input, img_input, img_input]) 
base_model = InceptionV3(input_tensor=img_conc, weights='imagenet', include_top=False)

for layer in base_model.layers:
    layer.trainable = False
    
x = Flatten()(base_model.output)
x = Dense(len(train_folders), activation='softmax')(x)

model = Model(inputs=base_model.input, outputs=x)

model.summary()
```

```python
#Rescaling the image from 48*48 to 150*150 for Transfer Learning in train set
import cv2
from PIL import Image
import numpy as np
folders_name_train = os.listdir(train_dataset)
print(folders_name_train)
for name in folders_name_train:
    for img in os.listdir(os.path.join(train_dataset,name)):
        image = Image.open(os.path.join(train_dataset,name,img))
        new_image = image.resize((150,150))
        new_image = np.array(new_image)
        cv2.imwrite(os.path.join(train_dataset,name,img), new_image)
```

```python
#Rescaling the image from 48*48 to 150*150 for Transfer Learning in test set
import cv2
from tqdm import tqdm
from PIL import Image
folders_name_test = os.listdir(test_dataset)
for name in tqdm(folders_name_test):
    for img in tqdm(os.listdir(os.path.join(test_dataset,name))):
        image = Image.open(os.path.join(test_dataset,name,img))
        new_image = image.resize((150,150))
        new_image = np.array(new_image)
        cv2.imwrite(os.path.join(test_dataset,name,img), new_image)
```

```python
model.compile(optimizer="adam", loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
train_datagen = ImageDataGenerator(rescale=1/255, shear_range=0.2, zoom_range=0.2, horizontal_flip=True,validation_split=0.15)
test_datagen = ImageDataGenerator(rescale=1/255)
```

```python
#Using flow from directory to locate the images for training
train_generator = train_datagen.flow_from_directory('Dataset/Train', target_size=(150,150),batch_size=32,color_mode='grayscale',class_mode="categorical")
test_generator = test_datagen.flow_from_directory('Dataset/Test', target_size=(150,150),batch_size=32,color_mode='grayscale',class_mode="categorical")
```

```python
history = model.fit_generator(
  train_generator,
  validation_data=test_generator,
  epochs=5,
  steps_per_epoch=len(train_generator),
  validation_steps=len(test_generator)
)
```

```python
model.save('model.h5')
```

```python
#Using Manual CNN
from keras.layers import MaxPooling2D
from keras.optimizers import Adam          
model1 = Sequential()

model1.add(Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(150,150,1)))
model1.add(MaxPooling2D(pool_size=(2, 2)))
model1.add(Dropout(0.25))

model1.add(Conv2D(64, kernel_size=(3, 3), activation='relu'))
model1.add(MaxPooling2D(pool_size=(2, 2)))
model1.add(Dropout(0.25))

model1.add(Conv2D(128, kernel_size=(3, 3), activation='relu'))
model1.add(MaxPooling2D(pool_size=(2, 2)))

model1.add(Conv2D(128, kernel_size=(3, 3), activation='relu'))
model1.add(MaxPooling2D(pool_size=(2, 2)))
model1.add(Dropout(0.25))

model1.add(Flatten())
model1.add(Dense(1024, activation='relu'))
model1.add(Dropout(0.5))
model1.add(Dense(len(train_folders), activation='softmax'))

model1.compile(loss='categorical_crossentropy',optimizer=Adam(lr=0.0001, decay=1e-6),metrics=['accuracy'])
```

```python
model_info = model1.fit_generator(
            train_generator,
            steps_per_epoch=len(train_generator),
            validation_steps=len(test_generator),
            epochs=5,
            validation_data=test_generator)
```

```python
model1.save('model1.h5')
```

```python
import matplotlib.pyplot as plt
plt.plot(model_info.history['loss'], label='train loss')
plt.plot(model_info.history['val_loss'], label='val loss')
plt.legend()
plt.show()
```

```python
plt.plot(model_info.history['accuracy'], label='train accuracy')
plt.plot(model_info.history['val_accuracy'], label='val accuracy')
plt.legend()
plt.show()
```
